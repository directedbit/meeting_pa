```python
import os
import re

recorded_file = ""
files = [f for f in os.listdir('.') if os.path.isfile(f)]
for f in files:
    if f.endswith('.mp3'):
      recorded_file = f
      print(f'using {recorded_file}')
      break
assert(recorded_file != "")
pattern = r"\d{4}-\d{2}-\d{2} .+\.mp3"

# Check if the file name matches the pattern
if re.match(pattern, recorded_file):
  print("Valid file name format.")
  os.environ['RECORDED_FILE'] = recorded_file
else:
  print("Invalid file format, failing..")
  assert(False)
```

```python
import os
import shutil
from pathlib import Path
import re


def convert_vtt_to_text(input_file, output_file):
    with open(input_file, 'r', encoding='utf-8') as file:
        lines = file.readlines()

    output_lines = []

    for line in lines:
        # Remove timestamp lines
        if re.match(r'.* --> .*', line):
            continue

        # Replace SPEAKER_00 with richard and SPEAKER_01 with gus
        line = line.replace('[SPEAKER_00]:', '[richard]:')
        line = line.replace('[SPEAKER_01]:', '[gus]:')

        # Remove blank lines
        if line.strip():
            output_lines.append(line.strip())

    with open(output_file, 'w', encoding='utf-8') as file:
        file.write("\n".join(output_lines))

base_folder = '/content/drive/MyDrive/Meetings'
base_filename = Path(recorded_file).stem
#base_filename = "2024-05-13 Test Meeting"
# Ensure the target folder exists
os.makedirs(base_folder, exist_ok=True)

# List of files to be renamed and moved
extensions = ['.mp3','.vtt']

# Loop through each extension, rename, and move the file
for ext in extensions:
    # Construct the source path for each file
    source_path = f'/content/{base_filename}{ext}'


    # Construct the new filename using the date and meeting name
    new_filename = base_filename.lower().replace(' ','-')+f'{ext}'
    if ext == '.vtt':
      new_text_file_name = base_filename.lower().replace(' ','-')+'.txt'
      convert_vtt_to_text(source_path, new_text_file_name)
      destination_file_name = os.path.join(base_folder, new_text_file_name)
      shutil.move(new_text_file_name, destination_file_name)
      print(f"Moved {new_text_file_name} to {destination_file_name}")

    # Create the destination path
    destination_path = os.path.join(base_folder, new_filename)
    print(f"moving {source_path} to {destination_path}")
    # Move the file with `shutil.move`
    if os.path.exists(source_path):
        shutil.move(source_path, destination_path)
        print(f"Moved: {source_path} to {destination_path}")
    else:
        print(f"File not found: {source_path}")
```
